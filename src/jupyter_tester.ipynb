```julia
using BentTool, StructuralUnits, Handcalcs, Plots
```

```julia
import BentTool: RectangularColumn, ColumnInfo, BarAInfo, BarBInfo, BarSInfo, BarTInfo, RectangularBentCap, ix, ix_rectangle, Bar
```

```julia
r = RectangularColumn(72, 96, 20);
col_info = ColumnInfo(r, 1, [19ft]);
bar_a = BarAInfo(Bar("#11"), 4, 12);
bar_b = BarBInfo(Bar("#11"), 4, 10);
bar_s = BarSInfo(Bar("#6"));
bar_t = BarTInfo(Bar("#6"), 6);
cap = RectangularBentCap(
	length= 38ft,
	width= 3.5ft,
	depth= 3.5ft,
	bar_a_info= bar_a,
	bar_b_info= bar_b,
	bar_s_info= bar_s,
	bar_t_info= bar_t,
	column_info= col_info,
	offset= 1ft,
);
```

```julia
plot(cap)
```

```julia
@handcalcs Ix = ix(cap)
```
